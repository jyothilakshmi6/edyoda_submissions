### 1) Implement Binary tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def inorder(self, node):
        if node is not None:
            self.inorder(node.left)
            print(node.val, end=' ')
            self.inorder(node.right)

    def preorder(self, node):
        if node is not None:
            print(node.val, end=' ')
            self.preorder(node.left)
            self.preorder(node.right)

    def postorder(self, node):
        if node is not None:
            self.postorder(node.left)
            self.postorder(node.right)
            print(node.val, end=' ')

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    print("Inorder traversal:")
    bt.inorder(bt.root)
    print("\nPreorder traversal:")
    bt.preorder(bt.root)
    print("\nPostorder traversal:")
    bt.postorder(bt.root)
```

### 2) Find height of a given tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def height(self, node):
        if node is None:
            return -1
        left_height = self.height(node.left)
        right_height = self.height(node.right)
        return 1 + max(left_height, right_height)

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    print("Height of the tree:", bt.height(bt.root))
```

### 5) Implement BFS (Breath First Search) and DFS (Depth First Search).

```python
from queue import Queue, LifoQueue

class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def bfs(self):
        if self.root is None:
            return
        q = Queue()
        q.put(self.root)
        while not q.empty():
            node = q.get()
            print(node.val, end=' ')
            if node.left is not None:
                q.put(node.left)
            if node.right is not None:
                q.put(node.right)

    def dfs(self):
        if self.root is None:
            return
        stack = LifoQueue()
        stack.put(self.root)
        while not stack.empty():
            node = stack.get()
            print(node.val, end=' ')
            if node.right is not None:
                stack.put(node.right)
            if node.left is not None:
                stack.put(node.left)

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    print("BFS traversal:")
    bt.bfs()
    print("\nDFS traversal:")
    bt.dfs()
```

### 6) Find sum of all left leaves in a given Binary Tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def sum_left_leaves(self, node):
        if node is None:
            return 0
        if node.left is not None and node.left.left is None and node.left.right is None:
            return node.left.val + self.sum_left_leaves(node.right)
        return self.sum_left_leaves(node.left) + self.sum_left_leaves(node.right)

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    print("Sum of all left leaves:", bt.sum_left_leaves(bt.root))
```

### 7) Find sum of all nodes of the given perfect binary tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def sum_nodes(self, node):
        if node is None:
            return 0
        return node.val + self.sum_nodes(node.left) + self.sum_nodes(node.right)

    def is_perfect(self, node):
        if node is None:
            return True
        if node.left is None and node.right is None:
            return True
        if node.left is not None and node.right is not None:
            return self.is_perfect(node.left) and self.is_perfect(node.right)
        return False

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    if bt.is_perfect(bt.root):
        print("Sum of all nodes:", bt.sum_nodes(bt.root))
    else:
        print("The tree is not a perfect binary tree.")
```

### 8) Count subtress that sum up to a given value x in a binary tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def count_subtrees(self, node, x):
        if node is None:
            return 0
        count = 0
        if node.left is not None:
            count += self.count_subtrees(node.left, x)
        if node.right is not None:
            count += self.count_subtrees(node.right, x)
        if node.val + (node.left.val if node.left is not None else 0) + (node.right.val if node.right is not None else 0) == x:
            count += 1
        return count

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    x = int(input("Enter the value of x: "))
    count = bt.count_subtrees(bt.root, x)
    print("Number of subtrees with sum", x, ":", count)
```

### 9) Find maximum level sum in Binary Tree.

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BinaryTree:
    def __init__(self):
        self.root = None

    def insert(self, val):
        if self.root is None:
            self.root = Node(val)
        else:
            self._insert(val, self.root)

    def _insert(self, val, node):
        if val < node.val:
            if node.left is None:
                node.left = Node(val)
            else:
                self._insert(val, node.left)
        else:
            if node.right is None:
                node.right = Node(val)
            else:
                self._insert(val, node.right)

    def max_level_sum(self):
        if self.root is None:
            return None
        max_sum = self.root.val
        queue = [self.root]
        while len(queue) > 0:
            level_sum = sum(node.val for node in queue)
            max_sum = max(max_sum, level_sum)
            next_level = []
            for node in queue:
                if node.left is not None:
                    next_level.append(node.left)
                if node.right is not None:
                    next_level.append(node.right)
            queue = next_level
        return max_sum

if __name__ == '__main__':
    bt = BinaryTree()
    n = int(input("Enter the number of nodes: "))
    print("Enter the values of the nodes:")
    for i in range(n):
        val = int(input())
        bt.insert(val)

    max_sum = bt.max_level_sum()
    print("Maximum level sum:", max_sum)
```
